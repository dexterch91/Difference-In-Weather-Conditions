```python
import pandas as pd

fileformat = "DAILYDATA_S24_"

#list of years of interest.
year = ["2020","2019","2018"] 

#dictionary to look up all available months for each year.
yearMonths = { 
    "2020":["01","02","03","04","05","06","07","08","09","10","11","12"],
    "2019":["01","02","03","04","05","06","07","08","09","10","11","12"],
    "2018":["01","02","03","04","05","06","07","08","09","10","11","12"]
}

#A lookup table for the area code in the script to look for the file
areaCode = {
    "Tengah":"S23",
    "Jurong West":"S44",
    "Changi":"S24"
}


def createDF(location,year,yearMonths):
    #Edit this to change location
    locCode = areaCode[location]
    fileformat = "DAILYDATA_"+locCode+"_"
    directory = ".\\data\\" + location + "\\" + fileformat
    
    for i in year:
        for j in yearMonths[i]:
            fileLink = directory + i+j +".csv"
            #print(fileLink)
            if (j=="01" and i=="2020"): #Create first dataframe to peg onto
                df1 = pd.read_csv(fileLink,na_values=["-","—",""," "],encoding= 'unicode_escape') #Create a dataframe from the list
                df1.columns = ['Station', 'Year', 'Month', 'Day', 'Daily Rainfall Total (mm)',
           'Highest 30 min Rainfall (mm)', 'Highest 60 min Rainfall (mm)',
           'Highest 120 min Rainfall (mm)', 'Mean Temperature (°C)',
           'Maximum Temperature (°C)', 'Minimum Temperature (°C)',
           'Mean Wind Speed (km/h)', 'Max Wind Speed (km/h)']
            else:
                df2 = pd.read_csv(fileLink,na_values=["-","—",""," "],encoding= 'unicode_escape');
                df2.columns = ['Station', 'Year', 'Month', 'Day', 'Daily Rainfall Total (mm)',
           'Highest 30 min Rainfall (mm)', 'Highest 60 min Rainfall (mm)',
           'Highest 120 min Rainfall (mm)', 'Mean Temperature (°C)',
           'Maximum Temperature (°C)', 'Minimum Temperature (°C)',
           'Mean Wind Speed (km/h)', 'Max Wind Speed (km/h)']
                df1 = pd.concat([df1,df2]); #Add on to the list
            #print(df)
        #numeric handling
        for i in df1.columns[1:]:
            df1[i] = pd.to_numeric(df1[i],errors="coerce")
    return df1 #retuning the dataframe

dfTengah = createDF("Tengah",year,yearMonths)
dfJurong = createDF("Jurong West",year,yearMonths)
dfChangi = createDF("Changi",year,yearMonths)
```

```python
import numpy as np

def mod_outlier(df):

        df2 = df["Mean Temperature (°C)"]._get_numeric_data()
        q1 = df["Mean Temperature (°C)"].quantile(0.25)
        q3 = df["Mean Temperature (°C)"].quantile(0.75)

        iqr = q3 - q1

        lower_bound = q1 -(1.5 * iqr) 
        upper_bound = q3 +(1.5 * iqr)
        print(lower_bound)
        print(upper_bound)

        for i in range(0,len(df2)):
            if df2[i] < lower_bound:            
                df2[i] = np.nan

            if df2[i] > upper_bound:            
                df2[i] = np.nan    

        return(df2)
    
dfTengah["Mean Temperature (°C)"]=mod_outlier(dfTengah)
dfJurong["Mean Temperature (°C)"]=mod_outlier(dfJurong)
dfChangi["Mean Temperature (°C)"]=mod_outlier(dfChangi)
dfTengah["Mean Temperature (°C)"]=mod_outlier(dfTengah)
```

```python
column_names = ['Station', 'Year', 'Month', 'Day', 'Daily Rainfall Total (mm)',
           'Highest 30 min Rainfall (mm)', 'Highest 60 min Rainfall (mm)',
           'Highest 120 min Rainfall (mm)', 'Mean Temperature (°C)',
           'Maximum Temperature (°C)', 'Minimum Temperature (°C)',
           'Mean Wind Speed (km/h)', 'Max Wind Speed (km/h)']
df_total = pd.DataFrame(columns = column_names) ##Creating new dataframe for total storage

for i in [dfChangi,dfJurong,dfTengah]:
    df_total = pd.concat([df_total,i])
    
df_total

df_total=df_total.reset_index()
df_total=df_total.drop(columns=['index'])
df_total
```

```python
# seed the pseudorandom number generator
from random import seed
from random import random
from random import randint
# seed random number generator
seed(1)
# generate some random numbers

def randDays(arrayDays,maxNum):
    arrayDays = []
    i = 0
    while (i < maxNum):
        r = randint(1,30) #generate random number
        if (r not in arrayDays): #ensure no duplicates
            arrayDays.append(r)
            i = i +1        
    return arrayDays

daysArray = []

sampleMonth = [9,8,9,8,9,8,9,9,8,9,8,9]

station = ["Changi","Jurong (West)","Tengah"]
df = pd.DataFrame(columns = column_names) ##Creating new dataframe for final storage

for i in range(12):
    monthArray = randDays(daysArray,sampleMonth[i]) #creating array of days
    daysArray.append(monthArray)
    
print(daysArray)
    
for location in station:
    df1 = df_total[df_total.Station == location]
    for i in range(12):
        df2 = df1[df1.Month == (i+1)] #add one to hit the right value for the month, starting with 1
        #Generate random numbers of interest for the month
        monthArray = daysArray[i]
        for j in monthArray:
            df3 = df2[df2.Day == j]
            df = pd.concat([df,df3]) #add to main dataset

df
```
